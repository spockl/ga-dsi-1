```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import wordcloud
%matplotlib inline
```

```python
color = wordcloud.get_single_color_func('deepskyblue')
```

```python
class_names = ['Sam', 'Joshua', 'Maria', 'Marina', 'Debjani', 'Tiana', 'Betsy', 'Kristen', 'Peter', 'Amish', 
              'Amer', 'Thomas', 'Doug', 'JP', 'Michael', 'Jesse', 'Andrew', 'Winston', 'Alex', 'Jocelyn', 'Sara',
              'Dyson', 'Rachel', 'Max', 'Tim']
```

```python
len(class_names)
```

```python
word = wordcloud.WordCloud(max_font_size=50, relative_scaling=0.3,\
                           background_color='white', color_func=color).generate(' '.join(class_names))
plt.figure(figsize=(16,8))
plt.imshow(word)
plt.axis("off")
plt.show()
```
