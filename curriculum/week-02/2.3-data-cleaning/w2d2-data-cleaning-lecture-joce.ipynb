## Common steps in cleaning data

- Drop outliers

For numerical data, consider dropping values > 3 SDs from the mean.

For categorical data, consider dropping cases accounting for < 1% of items.

- Normalize

Options include:

Max-min: $X_{norm} = (X − X_{min}) / (X_{max} − X_{min})$

Z-score: $(X_i - mean(X)) / sd$

## Common steps in cleaning data

- Relabel

*("Bachelor's", "BSc", "Bachelor of Arts") -> "BA"*

- Decode

*1 -> "EU", 2 -> "Asia-Pacific", 3 -> "MENA"*



## Common steps in cleaning 

- Recast

*("1.0", "2.0", "3.0") -> (1,2,3)*

- Handle null values

Pandas will usually impute NaNs for you. Drop them? Replace with estimates?

## Common steps in cleaning 

- Binarize (dummy variables)

*(Blue, Green, Blue, Red, Red, Green) ->*
*IsBlue: (1, 0, 1, 0, 0, 0); IsGreen: (0, 1, 0, 0, 0, 1); IsRed: (0, 0, 0, 1, 1, 0)*

- Discretization

*(20, 56, 7, 2, 14, 89, 70, 40) -> (Adult, Adult, Child, Child, Child, Senior, Senior, Adult)*

```python
sales = pd.read_csv('assets/datasets/sales_info.csv')
```

```python
sales.head()
```

```python
#Inspect the data types
sales.dtypes
```

```python
#Let's say all your values in the first column are too low by 1: use df.applymap to add 1 to each value in it
adjusted_sales = sales.copy()
adjusted_sales['volume_sold'] = adjusted_sales['volume_sold'].map(lambda x: x+1)
```

```python
#Use .value_counts to count the values of each column in the dataset
{x: adjusted_sales[x].value_counts() for x in adjusted_sales.columns.values}
```

```python
#Bonus
#Write functions to bin the numerical values in each column 
#into 'low', 'medium' and 'high' categories (hint: pandas 
#has a built-in quantile function)
def vol_cat(x):
    if x < adjusted_sales['volume_sold'].quantile(0.33):
        return 'low'
    elif x < adjusted_sales['volume_sold'].quantile(0.66):
        return 'medium'
    else:
        return 'high'
    
def margin_cat(x):
    if x < adjusted_sales['2015_margin'].quantile(0.33):
        return 'low'
    elif x < adjusted_sales['2015_margin'].quantile(0.66):
        return 'medium'
    else:
        return 'high'
    
def sales_2015_cat(x):
    if x < adjusted_sales['2015_q1_sales'].quantile(0.33):
        return 'low'
    elif x < adjusted_sales['2015_q1_sales'].quantile(0.66):
        return 'medium'
    else:
        return 'high'
    
def sales_2016_cat(x):
    if x < adjusted_sales['2016_q1_sales'].quantile(0.33):
        return 'low'
    elif x < adjusted_sales['2016_q1_sales'].quantile(0.66):
        return 'medium'
    else:
        return 'high'
```

```python
adjusted_sales['volume_sold_cat'] = adjusted_sales['volume_sold'].map(vol_cat)
adjusted_sales['2015_margin_cat'] = adjusted_sales['2015_margin'].map(margin_cat)
adjusted_sales['2015_q1_sales_cat'] = adjusted_sales['2015_q1_sales'].map(sales_2015_cat)
adjusted_sales['2016_q1_sales_cat'] = adjusted_sales['2016_q1_sales'].map(sales_2016_cat)
```

```python
adjusted_sales.head()
```

```python
#Use .value_counts again on each column of the dataset, and check that the ratios of the new values are what you expect
vals = {x: adjusted_sales[x].value_counts() for x in adjusted_sales.columns.values[4:]}
for k in vals:
    print k
    print vals[k]
    print ''
```
