```python
# Create a histogram by creating a list comprehension + a function to turn the data into inches
import string
def get_height_in_inches(i):
    if pd.isnull(i):
        return i
    elif 'Under' in i:
        return float(i[-5]) * 12 - 1
        #return 78.0
    elif 'above' in i:
        return float(i[0]) * 12 + float(i[2])
        #return 60.0
    else:
        return float(i[0]) * 12 + float(i[2:-1])
    #return [type(x) for x in lst]
```

```python
height_list = [get_height_in_inches(x) for x in char['height']]
```

```python
height_list[0:10]
```

```python
height_inch = char['height'].apply(get_height_in_inches)
```

```python
char['height'].apply(get_height_in_inches).value_counts()
```

```python
fig = height_inch.hist();
ax = plt.gca();
ax.set_title('Height distribution');
```
